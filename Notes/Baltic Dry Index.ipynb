```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tvDatafeed import TvDatafeed, Interval

tv = TvDatafeed()

bdi = pd.read_excel(
    '../data/Alter Data.xlsx',
    sheet_name = 'BDI',
    index_col = 0
)
bdi.index = pd.to_datetime(bdi.index)
```

```python
pan_ocean = tv.get_hist(
    symbol = '028670',
    exchange = 'KRX',
    interval = Interval.in_daily,
    n_bars=13000
)
pan_ocean.index = pd.to_datetime(pan_ocean.index.strftime('%Y-%m-%d'))
```

```python
bdi.loc['2020':, 'BDI'].plot(secondary_y=True)
pan_ocean.loc['2020':, 'close'].plot()
plt.show()
```
