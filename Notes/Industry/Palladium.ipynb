## Palladium: Industrial Linkages and Price Lead–Lag Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tvDatafeed import TvDatafeed, Interval

tv = TvDatafeed()
```

```python
palladium = tv.get_hist(
    symbol = 'PALLADIUM',
    exchange = 'TVC',
    interval = Interval.in_daily,
    n_bars = 13000
)
```

### Economic Role of Palladium

Palladium is primarily an industrial precious metal rather than a store-of-value commodity. Its demand structure is heavily concentrated in automotive emissions control systems, with additional usage in electronics, chemical catalysis, and emerging hydrogen-related applications. As a result, palladium prices tend to reflect industrial activity, regulatory pressure, and substitution dynamics more than monetary or inflationary factors.

```python
palladium['close'].plot()
plt.grid(False)
plt.show()
```

### Key Industries for Comparative Analysis

#### Automotive Industry (Primary Demand Driver)

The automotive sector, particularly gasoline-powered vehicles, accounts for the majority of palladium consumption through its use in three-way catalytic converters. Palladium demand is therefore closely linked to:

**Global gasoline vehicle production**

- Emission regulation stringency
- Autocatalyst loading intensity per vehicle

Because catalyst demand is derived demand, palladium prices often lead reported vehicle production data, especially during periods of tightening emission standards or supply disruptions.

**Relevant industry indicators**

- Global light vehicle production
- Gasoline vehicle sales (US, China, EU)
- Auto catalyst shipment volumes

### Platinum Substitution Dynamics

Platinum serves as the closest economic substitute for palladium in autocatalysts. When palladium prices rise significantly relative to platinum, automakers have incentives to re-optimize catalyst formulations.

This creates a medium-term feedback loop:
- Palladium price spikes → substitution R&D → gradual demand erosion
- Platinum price often lags palladium during substitution cycles

Key comparison
- Palladium–platinum price ratio
- Relative autocatalyst demand trends
